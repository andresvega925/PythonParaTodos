```python
class cuenta:
    def __init__(self,nombre, apellido,dni):
        self.total = 0
        self.nonbre = nombre
        self.appellido = apellido
        self.dni = dni        


class ahorro(cuenta):
        
    def consignar(self):        
        print("Ingrese el saldo total que va a consignar")
        val = float(input())
        self.total = self.total + val 
        
        return self.total
    
    def info_cuenta(self):      
       return "Nombre: ",self.nonbre,"Apellido: ",self.appellido," dni: ",self.dni
    
    def saldo_disponible(self):
        print("Su saldo disponible es: ",self.total)
        
         
    def retiro(self):
        print("¿Cuánto desea retirar?")
        val = float(input())
        
        if(val <= self.total and val > 0):
            print("Retiro exitoso")
            self.total = self.total - val
        else:
            pass


class principal:
    
    def ejecutar(self):
        pepo = ahorro("pepo","vega",1234)
        ejec = True
        while(ejec):
            print("[1] Consignar, \n[2] Retirar \n[3] Ver saldo \n[4] Info cuenta \n[5] salir")
            opr = int(input())
            if(opr == 1):
                pepo.consignar()
            if(opr == 2):
                pepo.retiro()
            elif(opr == 3):
                pepo.saldo_disponible()                    
            elif(opr == 4):
                print(pepo.info_cuenta())
            elif(opr == 5):
                ejec = False
            else:
                print("Opción no valida")


pr = principal()
pr.ejecutar()
    
```
