f(x) = e^$(-x^2)$

```python
def f(x):
    return np.exp(-x ** 2)
```

```python
x = np.linspace(-1,5, num=50)
x
```

```python
plt.plot(x,f(x),'or',label="Función f(x)")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.legend()
plt.title("Gráfica de la función f(x)")
```

## Gráficas orientadas a objetos

```python
fig, axes = plt.subplots()
axes.plot(x,f(x),'or', label="Función f(x)")
axes.set_xlim(-2,4)
axes.set_ylim(-1,2)
```

```python
fig, axes = plt.subplots(1,2, sharey=True)
axes[0].plot(x,f(x))
axes[0].set_xlabel("Eje x gráfica derecha")
axes[1].plot(x, -f(x),'r')
```

## Gráficas de funciones en dimensiones (Curvas de nivel)

```python
x = np.linspace(-2,2)
y = np.linspace(-2,2)

x2, y2 = np.meshgrid(x,y)
```


$g(x, y) = \cos(x) + \sin(y)^2$

```python
def g(x,y):
    return np.cos(x) * np.sin(y) ** 2
```

```python
z = g(x2,y2)
```

```python
fig, axes = plt.subplots()
axes.contour(x,y,z)
axes.contourf(x2,y2,z)
```
